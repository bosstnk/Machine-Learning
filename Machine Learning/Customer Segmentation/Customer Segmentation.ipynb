```python
# check outliers
outliers_perc = []

for col_name, col_data in customer_data[feature].items():
    # Column must be of numeric type (not object)
    if customer_data[col_name].dtype != 'O':  # 'O' object data type
        q1 = col_data.quantile(0.25)  # quartile 1 (25%)
        q3 = col_data.quantile(0.75)  # quartile 3 (75%)
        iqr = q3 - q1  # IQR (Interquartile Range)
        
        outliers = col_data[(col_data <= q1 - 1.5 * iqr) | (col_data >= q3 + 1.5 * iqr)]
        
        perc = len(outliers) * 100.0 / len(customer_data)
        outliers_perc.append((col_name, int(perc)))
        print(f"Column {col_name} outliers = {perc:.2f}%")
```

### Train Model

#### K-Means Clustering

Segmentation using two features: Annual Income and Spending Score.

```python
x = data_scaler[['Annual_Income', 'Spending_Score']].values
print(x)
```

find the number of clusters to optimum WCSS.

WCSS = Within Clusters Sum Square

```python
# finding wcss value for different number of clusters
wcss = [] # Create an empty list to hold many inertia values
silhoutte_scores = []
range_clusters = range(2,11)

for i in range_clusters:
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)

    cluster_labels = kmeans.fit_predict(x)
    sil_score = silhouette_score(x,cluster_labels)

    silhoutte_scores.append(sil_score)
    wcss.append(kmeans.inertia_)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))
# First Plot
ax1.plot(range_clusters, wcss, marker='o', linestyle='-')
ax1.set_xlabel('Number of Clusters')
ax1.set_ylabel('WCSS')
ax1.set_title("Elbow Method")
ax1.set_xticks(range(min(range_clusters), max(range_clusters) + 1))
ax1.grid(True)

# Second Plot
ax2.plot(range_clusters, silhoutte_scores, marker='o', color='orange')
ax2.set_xlabel('Number of Clusters')
ax2.set_ylabel('Silhoutte Score')
ax2.set_title('Silhoutte Score for Different Values of k')
ax2.set_xticks(range(min(range_clusters), max(range_clusters) + 1))
ax2.grid(True)

plt.tight_layout()
```

Optimum Number of Clusters = 5

```python
kmeans = KMeans(n_clusters = 5 ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 0  , algorithm='elkan') 

# return a label for each data point base on their cluster
cluster_labels = kmeans.fit_predict(x)
print(cluster_labels)
```

```python
data_modeling['Cluster'] = cluster_labels
```

```python
# plotting all the clusters and their Centroids
cluster_colors = {0: '#1f77b4', # Blue
                  1: '#ff7f0e', # Orange
                  2: '#2ca02c', # Green
                  3: '#d62728', # Red
                  4: '#5f236b', # Purple
}
plt.figure(figsize=(10,8))
colors = data_modeling['Cluster'].map(cluster_colors)

plt.scatter(x[:, 0], x[:, 1], c=colors, s=30,)

# plot the centroids
plt.scatter(x=kmeans.cluster_centers_[:,0], y=kmeans.cluster_centers_[:,1], s=50, c='cyan')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

#### Agglomerative Hierarchical Clustering

Creating a dendrogram using the three features (Age, Annual Income, Spending Score) of the scaled data set.

```python
data_scaler_np = data_scaler.values

plt.figure(figsize=(10,8))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
dendrogram = sch.dendrogram(sch.linkage(data_scaler_np, method = 'ward'))
plt.show()
```

- The dendrogram shows that the optimal number of clusters is between 4 and 6.

- According to the Elbow method in K-means clustering, the best number of clusters is 6.

- I will use 6 clusters to test the Hierarchical Clustering model against K-means.
